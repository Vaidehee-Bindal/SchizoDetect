# Apply 2nd Order Butterworth Filtering

EEG signals are weak (µV range) and easily contaminated by noise (muscle activity, eye movements, electrical interference). Filters are applied to extract relevant frequency bands (delta, theta, alpha, beta, gamma).
A 2nd-order Butterworth filter might be used for:
*   Low-pass filtering → remove high-frequency noise (e.g., >45 Hz)
*   High-pass filtering → remove slow drifts or baseline wander (<2.0 Hz)
*   Band-pass filtering → isolate EEG rhythms
*   Notch filtering → suppress 50/60 Hz powerline interference


```python
# 2nd Order Butterworth Bandpass Filter Function
def butter_bandpass_filter(data, lowcut, highcut, fs, order=2):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data)
    return y
```

```python
def plot_fft(signal, fs, title="FFT"):
    n = len(signal)
    f = np.fft.rfftfreq(n, d=1/fs)
    fft_magnitude = np.abs(np.fft.rfft(signal))

    plt.figure(figsize=(10, 4))
    plt.plot(f, fft_magnitude)
    plt.title(title)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()
```

```python
fs = 128  # Sampling Frequency in Hz

# First patient
patient = patients[0]        # shape (16, 7680)

# First channel
signal = patient[0, :]       # shape (7680,)

# Plot FFT before filtering
plot_fft(signal, fs, title="FFT Before Filtering (Raw EEG)")

# Apply Butterworth Filter (2–45 Hz)
filtered_signal = butter_bandpass_filter(signal, 2.0, 45.0, fs, order=2)

# Plot FFT after filtering
plot_fft(filtered_signal, fs, title="FFT After Filtering (2–45 Hz Butterworth)")
```

```python
def rereference_with_csd(signal_matrix, fs=128):
    # signal_matrix shape: (16 channels, timepoints)
    ch_names = ['F7', 'F3', 'F4', 'F8', 'T3', 'C3', 'Cz', 'C4', 'T4', 'T5', 'P3', 'Pz', 'P4', 'T6', 'O1', 'O2']

    info = mne.create_info(ch_names=ch_names, sfreq=fs, ch_types='eeg')

    # Average Reference
    avg_ref = signal_matrix - np.mean(signal_matrix, axis=0, keepdims=True)

    # Standard Reference to Cz (Index 6)
    cz_ref = avg_ref - avg_ref[6:7, :]

    # Apply CSD (Current Source Density)
    raw = mne.io.RawArray(cz_ref * 1e-6, info)  # µV to V for MNE
    montage = mne.channels.make_standard_montage('standard_1020')
    raw.set_montage(montage)
    raw_csd = compute_current_source_density(raw)
    csd_data = raw_csd.get_data() * 1e6  # Convert back to µV

    return csd_data
```

```python
def save_scalogram_for_cnn(patient_id, channel_id, segment_idx, signal, save_root, fs=128):
    bandwidth = 3.0
    wavelet = fcwt.Morlet(bandwidth)

    f0, f1, fn = 2.0, 45.0, 100
    scales = fcwt.Scales(wavelet, 0, fs, f0, f1, fn)
    transform = fcwt.FCWT(wavelet, 1, True, True)

    output = np.empty((scales.nscales, len(signal)), dtype=np.complex64)
    transform.cwt(signal.astype(np.float32), scales, output)
    scalogram = np.abs(output)

    save_dir = os.path.join(save_root, f"P{patient_id}/Ch{channel_id}")
    os.makedirs(save_dir, exist_ok=True)
    save_path = os.path.join(save_dir, f"Seg_{segment_idx}.png")

    if os.path.exists(save_path):
        print(f"Skipping existing: {save_path}")
        return

    fig, ax = plt.subplots(figsize=(5, 5), dpi=200)
    cax = ax.imshow(
        scalogram,
        extent=[0, len(signal)/fs, f0, f1],
        cmap='jet',
        origin='lower',
        interpolation='bicubic',
        aspect='auto'
    )
    ax.axis('off')
    fig.savefig(save_path, bbox_inches='tight', pad_inches=0, dpi=200)
    plt.close(fig)

    img = Image.open(save_path).resize((256, 256))
    img.save(save_path)
```

```python
save_root = "/content/drive/MyDrive/EEG_Project/Scalograms_ReReferenced"
fs = 128
window_size = 20 * fs  # 20 seconds
step_size = 5 * fs     # 5 seconds

for idx, patient in enumerate(patients):
    # Re-reference + CSD
    patient_reref = rereference_with_csd(patient, fs=fs)

    for ch in range(patient_reref.shape[0]):  # 16 channels
        signal = patient_reref[ch, :]

        # utterworth Filter (2-45 Hz)
        filtered_signal = butter_bandpass_filter(signal, 2.0, 45.0, fs, order=2)

        # Segmenting into Windows + Save Scalograms
        start_points = np.arange(0, len(filtered_signal) - window_size + 1, step_size)

        for win_idx, start in enumerate(start_points):
            segment = filtered_signal[start:start + window_size]
            save_scalogram_for_cnn(
                patient_id=idx + 1,
                channel_id=ch + 1,
                segment_idx=win_idx + 1,
                signal=segment,
                save_root=save_root,
                fs=fs
            )

print("\nAll filtered, re-referenced, CSD-applied scalograms saved.")
```

# Applying ABSO and Creating Train-Test Data file

```python
# Create output directory
output_dir = '/content/drive/MyDrive/EEG_Project/Classification_Results'
os.makedirs(output_dir, exist_ok=True)

# Checkpoint paths
asbo_checkpoint = os.path.join(output_dir, 'asbo_selected_features.pkl')
train_data_checkpoint = os.path.join(output_dir, 'train_data.pkl')
test_data_checkpoint = os.path.join(output_dir, 'test_data.pkl')

def load_features():
    try:
        with open('/content/drive/MyDrive/EEG_Project/densenet169_features.pkl', 'rb') as f:
            densenet_features_dict = pickle.load(f)
        with open('/content/drive/MyDrive/EEG_Project/efficientnetb3_features.pkl', 'rb') as f:
            efficientnet_features_dict = pickle.load(f)

        print(f"Total DenseNet169 images: {len(densenet_features_dict)}")
        print(f"Total EfficientNetB3 images: {len(efficientnet_features_dict)}")

        common_paths = sorted(set(densenet_features_dict.keys()) & set(efficientnet_features_dict.keys()))
        if not common_paths:
            raise ValueError("No common image paths found in both feature dictionaries.")
        print(f"Common images: {len(common_paths)}")
        print(f"Sample common paths: {common_paths[:5]}")

        subject_ids = [path.split('/')[-3] for path in common_paths]
        unique_subjects = sorted(set(subject_ids))
        print(f"Unique subjects: {len(unique_subjects)}")
        print(f"Sample subject IDs: {subject_ids[:5]}")
        if len(unique_subjects) != 84:
            raise ValueError(f"Expected 84 subjects, got {len(unique_subjects)}")

        subject_image_counts = Counter(subject_ids)
        print("\nImages per subject:")
        for subject, count in sorted(subject_image_counts.items()):
            print(f"Subject {subject}: {count} images")
            if count != 144:
                raise ValueError(f"Subject {subject} has {count} images, expected 144")

        total_images = sum(subject_image_counts.values())
        print(f"Total images: {total_images}")
        if total_images != 12096:
            raise ValueError(f"Total images {total_images}, expected 12096")

        densenet_features = np.array([densenet_features_dict[path] for path in common_paths])
        efficientnet_features = np.array([efficientnet_features_dict[path] for path in common_paths])

        print(f"DenseNet169 features shape: {densenet_features.shape}")
        print(f"EfficientNetB3 features shape: {efficientnet_features.shape}")

        if densenet_features.ndim != 2 or efficientnet_features.ndim != 2:
            raise ValueError("Feature arrays must be 2D after conversion.")
        if densenet_features.shape[0] != efficientnet_features.shape[0]:
            raise ValueError(f"Mismatched sample sizes: DenseNet169 has {densenet_features.shape[0]} samples, "
                            f"EfficientNetB3 has {efficientnet_features.shape[0]} samples.")
        if densenet_features.shape[1] != 1664:
            raise ValueError(f"Unexpected DenseNet169 feature dimension: {densenet_features.shape[1]}, expected 1664.")
        if efficientnet_features.shape[1] != 1536:
            raise ValueError(f"Unexpected EfficientNetB3 feature dimension: {efficientnet_features.shape[1]}, expected 1536.")

        return densenet_features, efficientnet_features, common_paths, subject_ids, subject_image_counts
    except Exception as e:
        print(f"Error loading feature files: {e}")
        raise

def concatenate_features(densenet_features, efficientnet_features):
    try:
        concatenated_features = np.concatenate((efficientnet_features, densenet_features), axis=1)
        print(f"Concatenated features shape: {concatenated_features.shape}")
        return concatenated_features
    except Exception as e:
        print(f"Error concatenating features: {e}")
        print(f"DenseNet169 features shape: {densenet_features.shape}")
        print(f"EfficientNetB3 features shape: {efficientnet_features.shape}")
        raise

def asbo_feature_selection(features, labels, max_iterations=200, population_size=50, checkpoint_path=asbo_checkpoint):
    if np.any(np.isnan(features)) or np.any(np.isinf(features)):
        raise ValueError("Features contain NaN or infinite values")
    if features.shape[0] != len(labels):
        raise ValueError(f"Feature samples ({features.shape[0]}) mismatch with labels ({len(labels)})")

    if os.path.exists(checkpoint_path):
        print("Loading ASBO selected features from checkpoint...")
        with open(checkpoint_path, 'rb') as f:
            selected_features = pickle.load(f)
        print(f"Loaded {np.sum(selected_features)} selected features from checkpoint")
        return selected_features

    n_features = features.shape[1]
    population = np.random.randint(0, 2, size=(population_size, n_features))

    def fitness(feature_subset, features, labels):
        selected_features = features[:, feature_subset == 1]
        if selected_features.shape[1] == 0:
            return 0
        X_train, X_test, y_train, y_test = train_test_split(
            selected_features, labels, test_size=0.2, random_state=42
        )
        knn = KNeighborsClassifier(n_neighbors=7, weights='distance')
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        return f1_score(y_test, y_pred, average='weighted')

    best_fitness = -1
    best_solution = None

    for iteration in tqdm(range(max_iterations), desc="ASBO Iterations"):
        fitness_scores = np.array(Parallel(n_jobs=-1)(
            delayed(fitness)(ind, features, labels) for ind in population
        ))
        best_idx = np.argmax(fitness_scores)
        worst_idx = np.argmin(fitness_scores)
        best_solution_iter = population[best_idx]
        worst_solution_iter = population[worst_idx]

        if fitness_scores[best_idx] > best_fitness:
            best_fitness = fitness_scores[best_idx]
            best_solution = best_solution_iter.copy()

        avg_solution = (best_solution_iter + worst_solution_iter) / 2
        avg_solution = (sigmoid(avg_solution) > np.random.uniform(0, 1, avg_solution.shape)).astype(int)
        diff_solution = best_solution_iter - worst_solution_iter
        diff_solution = (sigmoid(diff_solution) > np.random.uniform(0, 1, diff_solution.shape)).astype(int)
        exploit_solution = best_solution_iter + np.random.uniform(-0.1, 0.1, best_solution_iter.shape)
        exploit_solution = (sigmoid(exploit_solution) > np.random.uniform(0, 1, exploit_solution.shape)).astype(int)

        worst_indices = np.argsort(fitness_scores)[:3]
        population[worst_indices[0]] = avg_solution
        population[worst_indices[1]] = diff_solution
        population[worst_indices[2]] = exploit_solution

    print(f"Selected {np.sum(best_solution)} features out of {n_features}")
    with open(checkpoint_path, 'wb') as f:
        pickle.dump(best_solution, f)
    return best_solution

def sigmoid(x):
    x = np.clip(x, -500, 500)
    return 1 / (1 + np.exp(-x))

def main():
    densenet_features, efficientnet_features, common_paths, subject_ids, subject_image_counts = load_features()
    print(f"Number of common paths: {len(common_paths)}")
    expected_samples = 12096
    if len(common_paths) != expected_samples:
        raise ValueError(f"Expected {expected_samples} samples, got {len(common_paths)}. Check feature dictionaries for missing entries.")

    main.subject_image_counts = subject_image_counts

    global labels
    print(f"Original labels type: {type(labels)}, length: {len(labels)}")
    print(f"Original labels sample: {labels[:5]}")
    if isinstance(labels, np.ndarray):
        print(f"Original labels shape: {labels.shape}")
        print(f"Original labels dtype: {labels.dtype}")

    unique_subjects = sorted(set(subject_ids))
    if isinstance(labels, (list, np.ndarray)):
        if len(labels) != 84:
            raise ValueError(f"Expected 84 labels, got {len(labels)}")
        labels = np.array(labels, dtype=np.int32)
        if labels.ndim > 1:
            print(f"Warning: Original labels are {labels.ndim}D with shape {labels.shape}. Converting to 1D.")
            if labels.ndim == 2 and labels.shape[1] > 1 and np.all((labels == 0) | (labels == 1)):
                print("Labels are one-hot encoded. Converting with argmax.")
                labels = np.argmax(labels, axis=1)
            else:
                print(f"Labels not one-hot encoded. Taking first column.")
                labels = labels[:, 0]
        subject_label_map = {subj: lbl for subj, lbl in zip(unique_subjects, labels)}
    elif isinstance(labels, dict):
        if len(labels) != 84:
            raise ValueError(f"Expected 84 labels, got {len(labels)}")
        try:
            labels = np.array([labels[subj] for subj in unique_subjects], dtype=np.int32)
        except KeyError:
            raise ValueError("Label dictionary keys do not match subject IDs")
        if labels.ndim > 1:
            print(f"Warning: Label values are {labels.ndim}D with shape {labels.shape}. Converting to 1D.")
            if labels.ndim == 2 and labels.shape[1] > 1 and np.all((labels == 0) | (labels == 1)):
                print("Label values are one-hot encoded. Converting with argmax.")
                labels = np.argmax(labels, axis=1)
            else:
                print(f"Label values not one-hot encoded. Taking first column.")
                labels = labels[:, 0]
        subject_label_map = {subj: lbl for subj, lbl in zip(unique_subjects, labels)}
    else:
        raise ValueError(f"Unsupported labels type: {type(labels)}. Must be dict, list, or numpy array.")

    unique_labels, counts = np.unique(labels, return_counts=True)
    print(f"Original label distribution: {dict(zip(unique_labels, counts))}")
    if counts[0] != 39 or counts[1] != 45:
        raise ValueError(f"Expected 39 healthy (0) and 45 schizophrenic (1), got {counts[0]} healthy and {counts[1]} schizophrenic")

    healthy_subjects = [subj for subj, lbl in subject_label_map.items() if lbl == 0]
    schizo_subjects = [subj for subj, lbl in subject_label_map.items() if lbl == 1]
    print(f"Healthy subjects: {len(healthy_subjects)}")
    print(f"Schizophrenic subjects: {len(schizo_subjects)}")
    if len(healthy_subjects) != 39 or len(schizo_subjects) != 45:
        raise ValueError(f"Expected 39 healthy and 45 schizophrenic subjects, got {len(healthy_subjects)} healthy and {len(schizo_subjects)} schizophrenic")

    random.seed(42)
    train_healthy = random.sample(healthy_subjects, 33)
    train_schizo = random.sample(schizo_subjects, 34)
    train_subjects = train_healthy + train_schizo
    random.shuffle(train_subjects)
    test_subjects = [subj for subj in unique_subjects if subj not in train_subjects]
    random.shuffle(test_subjects)

    # Validate train-test split
    train_healthy_count = sum(subject_label_map[subj] == 0 for subj in train_subjects)
    train_schizo_count = sum(subject_label_map[subj] == 1 for subj in train_subjects)
    test_healthy_count = sum(subject_label_map[subj] == 0 for subj in test_subjects)
    test_schizo_count = sum(subject_label_map[subj] == 1 for subj in test_subjects)
    print(f"Training subjects: {len(train_subjects)} (Healthy: {train_healthy_count}, Schizophrenic: {train_schizo_count})")
    print(f"Training subject IDs: {train_subjects}")
    print(f"Test subjects: {len(test_subjects)} (Healthy: {test_healthy_count}, Schizophrenic: {test_schizo_count})")
    print(f"Test subject IDs: {test_subjects}")
    if train_healthy_count != 33 or train_schizo_count != 34:
        raise ValueError(f"Expected 33 healthy and 34 schizophrenic in training set, got {train_healthy_count} healthy and {train_schizo_count} schizophrenic")
    if test_healthy_count != 6 or test_schizo_count != 11:
        raise ValueError(f"Expected 6 healthy and 11 schizophrenic in test set, got {test_healthy_count} healthy and {test_schizo_count} schizophrenic")

    train_image_count = sum(subject_image_counts[subj] for subj in train_subjects)
    test_image_count = sum(subject_image_counts[subj] for subj in test_subjects)
    print(f"Training images: {train_image_count}")
    print(f"Test images: {test_image_count}")

    train_indices = [i for i, subj in enumerate(subject_ids) if subj in train_subjects]
    test_indices = [i for i, subj in enumerate(subject_ids) if subj in test_subjects]
    concatenated_features = concatenate_features(densenet_features, efficientnet_features)
    X_train = concatenated_features[train_indices]
    X_test = concatenated_features[test_indices]

    y_train = []
    for subj in train_subjects:
        y_train.extend([subject_label_map[subj]] * subject_image_counts[subj])
    y_train = np.array(y_train, dtype=np.int32)

    y_test = []
    for subj in test_subjects:
        y_test.extend([subject_label_map[subj]] * subject_image_counts[subj])
    y_test = np.array(y_test, dtype=np.int32)

    print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
    print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
    print(f"y_train distribution: {dict(zip(*np.unique(y_train, return_counts=True)))}")
    print(f"y_test distribution: {dict(zip(*np.unique(y_test, return_counts=True)))}")
    print(f"y_train sample: {y_train[:5]}")
    print(f"y_test sample: {y_test[:5]}")

    # Save training and testing data
    with open(train_data_checkpoint, 'wb') as f:
        pickle.dump({'X_train': X_train, 'y_train': y_train}, f)
    print(f"Saved training data to {train_data_checkpoint}")

    with open(test_data_checkpoint, 'wb') as f:
        pickle.dump({'X_test': X_test, 'y_test': y_test}, f)
    print(f"Saved testing data to {test_data_checkpoint}")

    start_idx = 0
    for subj in train_subjects:
        count = subject_image_counts[subj]
        block = y_train[start_idx:start_idx + count]
        if len(np.unique(block)) > 1:
            raise ValueError(f"Training block for subject {subj} (indices {start_idx}:{start_idx + count}) contains multiple labels")
        else:
            print(f"Confirmed: Training block for subject {subj} (indices {start_idx}:{start_idx + count}) all equal to {block[0]}")
        start_idx += count

    start_idx = 0
    for subj in test_subjects:
        count = subject_image_counts[subj]
        block = y_test[start_idx:start_idx + count]
        if len(np.unique(block)) > 1:
            raise ValueError(f"Test block for subject {subj} (indices {start_idx}:{start_idx + count}) contains multiple labels")
        else:
            print(f"Confirmed: Test block for subject {subj} (indices {start_idx}:{start_idx + count}) all equal to {block[0]}")
        start_idx += count

    train_hashes = [hash(tuple(x)) for x in X_train]
    test_hashes = [hash(tuple(x)) for x in X_test]
    if any(h in train_hashes for h in test_hashes):
        raise ValueError("Data leakage detected: Overlap between training and test sets!")

    X_train = X_train.astype(np.float32)
    selected_features = asbo_feature_selection(X_train, y_train, max_iterations=200, population_size=50)

if __name__ == "__main__":
    # Define labels: 39 healthy (0), 45 schizophrenic (1)
    labels = np.array([0] * 39 + [1] * 45, dtype=np.int32)  # 39 healthy, 45 schizophrenic
    output_dir = '/content/drive/MyDrive/EEG_Project/Classification_Results'
    os.makedirs(output_dir, exist_ok=True)
    main()
```

```python
import numpy as np
import pandas as pd
import pickle
import os
import hashlib
import tempfile
import shutil
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, KFold, ParameterGrid
from deap import base, creator, tools, algorithms
from tqdm import tqdm
import signal
import sys
import joblib

# Signal handler for graceful interruption
def signal_handler(sig, frame):
    print("\nInterruption detected. Saving current state...")
    save_checkpoint(current_state)
    sys.exit(0)

signal.signal(signal.SIGINT, signal_handler)

# Compute MD5 hash for file integrity
def compute_md5(file_path):
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()

# Save checkpoint with atomic write
def save_checkpoint(state):
    checkpoint_dir = "/content/drive/MyDrive/EEG_Project/Classification_Results/Checkpoints/"
    os.makedirs(checkpoint_dir, exist_ok=True)
    temp_fd, temp_path = tempfile.mkstemp(suffix='.pkl', dir=checkpoint_dir)

    try:
        with os.fdopen(temp_fd, 'wb') as f:
            pickle.dump(state, f)
        checkpoint_path = os.path.join(checkpoint_dir, f"knn_ga_checkpoint_iter_{state.get('iteration', 0)}.pkl")
        shutil.move(temp_path, checkpoint_path)
        print(f"Checkpoint saved at iteration {state.get('iteration', 0)}: {checkpoint_path}")
    except Exception as e:
        print(f"Error saving checkpoint: {e}")
        if os.path.exists(temp_path):
            os.remove(temp_path)
        raise

# Load checkpoint with integrity check
def load_checkpoint():
    checkpoint_dir = "/content/drive/MyDrive/EEG_Project/Classification_Results/Checkpoints/"
    if not os.path.exists(checkpoint_dir):
        return None
    checkpoint_files = [f for f in os.listdir(checkpoint_dir) if f.startswith("knn_ga_checkpoint_iter_")]
    if not checkpoint_files:
        return None

    latest_checkpoint = max(checkpoint_files, key=lambda x: int(x.split('_iter_')[1].split('.pkl')[0]))
    checkpoint_path = os.path.join(checkpoint_dir, latest_checkpoint)

    try:
        md5_before = compute_md5(checkpoint_path)
        with open(checkpoint_path, 'rb') as f:
            state = pickle.load(f)
        md5_after = hashlib.md5(open(checkpoint_path, 'rb').read()).hexdigest()
        if md5_before != md5_after:
            print(f"Checkpoint corrupted: {checkpoint_path}")
            return None
        print(f"Loaded checkpoint: {checkpoint_path}")
        return state
    except Exception as e:
        print(f"Error loading checkpoint: {e}")
        return None

# Load and concatenate features without PCA
def load_and_concatenate_features():
    densenet_path = "/content/drive/MyDrive/EEG_Project/densenet169_features.pkl"
    efficientnet_path = "/content/drive/MyDrive/EEG_Project/efficientnetb3_features.pkl"

    with open(densenet_path, 'rb') as f:
        densenet_data = pickle.load(f)
    with open(efficientnet_path, 'rb') as f:
        efficientnet_data = pickle.load(f)

    if not isinstance(densenet_data, dict) or not isinstance(efficientnet_data, dict):
        raise ValueError(f"Expected dictionaries. Got DenseNet type: {type(densenet_data)}, EfficientNet type: {type(efficientnet_data)}")

    expected_samples = 12096
    densenet_keys = sorted(densenet_data.keys())
    efficientnet_keys = sorted(efficientnet_data.keys())

    if len(densenet_keys) != expected_samples or len(efficientnet_keys) != expected_samples:
        raise ValueError(f"Expected {expected_samples} samples, got {len(densenet_keys)} for DenseNet and {len(efficientnet_keys)} for EfficientNet")
    if densenet_keys != efficientnet_keys:
        raise ValueError("Image paths do not match")

    densenet_features = np.array([densenet_data[key] for key in densenet_keys])
    efficientnet_features = np.array([efficientnet_data[key] for key in efficientnet_keys])

    if densenet_features.shape != (12096, 1664):
        raise ValueError(f"Expected X_train shape (12096, 1664), got {densenet_features.shape}")
    if efficientnet_features.shape != (12096, 1536):
        raise ValueError(f"Expected X_test shape (12096, 1536), got {efficientnet_features.shape}")

    concatenated_features = np.concatenate((densenet_features, efficientnet_features), axis=1)
    print(f"Concatenated features shape: {concatenated_features.shape}")

    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/train_data.pkl", 'rb') as f:
        train_data = pickle.load(f)
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/test_data.pkl", 'rb') as f:
        test_data = pickle.load(f)

    y_train = np.array(train_data['y_train'])
    y_test = np.array(test_data['y_test'])

    if y_train.shape != (9648,) or y_test.shape != (2448,):
        raise ValueError(f"Expected y_train shape (9648,), y_test shape (2448,), got {y_train.shape}, {y_test.shape}")
    if not np.all(np.isin(y_train, [0, 1])) or not np.all(np.isin(y_test, [0, 1])):
        raise ValueError("Labels must be binary (0 or 1)")

    scaler = StandardScaler()
    X_train = scaler.fit_transform(concatenated_features[:9648])
    X_test = scaler.transform(concatenated_features[9648:12096])

    print(f"Training label distribution: {np.bincount(y_train)}")
    print(f"Testing label distribution: {np.bincount(y_test)}")

    return X_train, y_train, X_test, y_test

# Compute metrics and confusion matrix
def compute_metrics(y_true, y_pred, dataset_name, iteration=None, total_iterations=None, params=None):
    cm = confusion_matrix(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    if iteration is not None and total_iterations is not None:
        print(f"\nIteration {iteration}/{total_iterations} - Params: {params}, {dataset_name}:")
    else:
        print(f"\n{dataset_name} Metrics:")
    print(f"Confusion Matrix:\n{cm}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")

    return {
        'confusion_matrix': cm,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    }

# Genetic Algorithm for feature selection with detailed metrics
def evaluate_features(individual, X_train, y_train, cv, default_knn_params, gen, n_generations):
    selected_features = [i for i, bit in enumerate(individual) if bit]
    if not selected_features:
        return 0.0,

    X_train_selected = X_train[:, selected_features]
    knn = KNeighborsClassifier(**default_knn_params)
    scores = []
    precisions = []
    recalls = []
    f1s = []

    for fold, (train_idx, val_idx) in enumerate(cv.split(X_train)):
        X_t, X_v = X_train_selected[train_idx], X_train_selected[val_idx]
        y_t, y_v = y_train[train_idx], y_train[val_idx]
        knn.fit(X_t, y_t)
        y_pred = knn.predict(X_v)
        metrics = compute_metrics(y_v, y_pred, f"Generation {gen+1} Fold {fold+1} Validation",
                                 iteration=f"{gen+1}.{fold+1}", total_iterations=f"{n_generations}.5",
                                 params=f"Features: {len(selected_features)}")
        scores.append(metrics['accuracy'])
        precisions.append(metrics['precision'])
        recalls.append(metrics['recall'])
        f1s.append(metrics['f1_score'])

    return np.mean(scores),

# Feature selection using Genetic Algorithm
def genetic_algorithm_feature_selection(X_train, y_train, n_features=3200, n_selected_features=1000, n_generations=10, population_size=20):
    global current_state
    checkpoint = load_checkpoint()
    if checkpoint and 'best_features' in checkpoint:
        print("Resuming with previously selected features...")
        return checkpoint['best_features']

    creator.create("FitnessMax", base.Fitness, weights=(1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMax)

    toolbox = base.Toolbox()
    toolbox.register("attr_bool", np.random.choice, [0, 1], p=[1 - n_selected_features/n_features, n_selected_features/n_features])
    toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, n=n_features)
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)
    toolbox.register("evaluate", evaluate_features, X_train=X_train, y_train=y_train,
                     cv=KFold(n_splits=5, shuffle=True, random_state=42),
                     default_knn_params={'n_neighbors': 5, 'weights': 'uniform', 'metric': 'euclidean'},
                     n_generations=n_generations)
    toolbox.register("mate", tools.cxTwoPoint)
    toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
    toolbox.register("select", tools.selTournament, tournsize=3)

    population = toolbox.population(n=population_size)
    best_individual = None
    best_fitness = -float('inf')

    for gen in tqdm(range(n_generations), desc="GA Progress"):
        if checkpoint and 'ga_completed_generations' in checkpoint and gen < checkpoint['ga_completed_generations']:
            print(f"Skipping completed generation {gen+1}/{n_generations}")
            population = checkpoint['ga_population']
            continue

        offspring = algorithms.varAnd(population, toolbox, cxpb=0.5, mutpb=0.2)
        fits = [toolbox.evaluate(ind, gen=gen) for ind in offspring]

        for ind, fit in zip(offspring, fits):
            ind.fitness.values = fit

        population = toolbox.select(offspring, k=len(population))
        top_individual = tools.selBest(population, k=1)[0]
        top_fitness = top_individual.fitness.values[0]

        print(f"Generation {gen+1}/{n_generations} - Best Fitness (Accuracy): {top_fitness:.4f}")

        if top_fitness > best_fitness:
            best_fitness = top_fitness
            best_individual = top_individual

        current_state['ga_completed_generations'] = gen + 1
        current_state['ga_population'] = population
        current_state['best_features'] = [i for i, bit in enumerate(best_individual) if bit]
        save_checkpoint(current_state)

    selected_features = [i for i, bit in enumerate(best_individual) if bit]
    print(f"Selected {len(selected_features)} features")
    return selected_features

# Custom GridSearchCV with explicit parameters and progress bar
class CustomGridSearchCV(GridSearchCV):
    def __init__(self, estimator, param_grid, scoring=None, n_jobs=None, refit=True, cv=None,
                 verbose=0, pre_dispatch='2*n_jobs', error_score=np.nan, return_train_score=False):
        super().__init__(
            estimator=estimator,
            param_grid=param_grid,
            scoring=scoring,
            n_jobs=n_jobs,
            refit=refit,
            cv=cv,
            verbose=verbose,
            pre_dispatch=pre_dispatch,
            error_score=error_score,
            return_train_score=return_train_score
        )
        self.current_state = {
            'iteration': 0,
            'completed_params': [],
            'cv_results': {'params': [], 'mean_test_score': [], 'std_test_score': [],
                           'mean_precision': [], 'mean_recall': [], 'mean_f1': []}
        }

    def fit(self, X, y):
        global current_state
        checkpoint = load_checkpoint()
        if checkpoint and 'tuning_state' in checkpoint:
            self.current_state = checkpoint['tuning_state']
            completed_params = set(tuple(sorted((k, v) for k, v in d.items())) for d in checkpoint['tuning_state']['completed_params'])
        else:
            completed_params = set()

        # Explicitly set joblib backend to multiprocessing
        with joblib.parallel_backend('multiprocessing'):
            param_combinations = list(ParameterGrid(self.param_grid))
            total_iterations = len(param_combinations) * self.cv
            iteration = self.current_state['iteration']

            for param_idx, params in tqdm(enumerate(param_combinations), total=len(param_combinations), desc="GridSearchCV Params"):
                param_tuple = tuple(sorted((k, v) for k, v in params.items()))
                if param_tuple in completed_params:
                    print(f"Skipping completed parameters: {params}")
                    continue

                print(f"\nEvaluating parameters: {params}")
                knn = KNeighborsClassifier(**params, n_jobs=1)  # Disable n_jobs=-1 inside parallel loop
                cv = KFold(n_splits=self.cv, shuffle=True, random_state=42)

                scores = []
                precisions = []
                recalls = []
                f1s = []
                for fold, (train_idx, val_idx) in enumerate(cv.split(X)):
                    iteration += 1
                    X_train, X_val = X[train_idx], X[val_idx]
                    y_train, y_val = y[train_idx], y[val_idx]

                    knn.fit(X_train, y_train)
                    y_pred = knn.predict(X_val)
                    metrics = compute_metrics(y_val, y_pred, f"Fold {fold+1} Validation",
                                             iteration=iteration, total_iterations=total_iterations, params=params)

                    scores.append(metrics['accuracy'])
                    precisions.append(metrics['precision'])
                    recalls.append(metrics['recall'])
                    f1s.append(metrics['f1_score'])

                    self.current_state['iteration'] = iteration
                    self.current_state['cv_results']['params'].append(params)
                    self.current_state['cv_results']['mean_test_score'].append(np.mean(scores))
                    self.current_state['cv_results']['std_test_score'].append(np.std(scores))
                    self.current_state['cv_results']['mean_precision'].append(np.mean(precisions))
                    self.current_state['cv_results']['mean_recall'].append(np.mean(recalls))
                    self.current_state['cv_results']['mean_f1'].append(np.mean(f1s))
                    current_state['tuning_state'] = self.current_state
                    save_checkpoint(current_state)

                self.current_state['completed_params'].append(params)
                completed_params.add(param_tuple)

            super().fit(X, y)
        return self

# Main classifier function with GA feature selection and kNN tuning
def knn_ga_classifier(X_train, y_train, X_test, y_test):
    global current_state
    # Compute class weights to handle imbalance
    class_weights = {0: len(y_train) / (2 * np.bincount(y_train)[0]),
                     1: len(y_train) / (2 * np.bincount(y_train)[1])}

    # Feature selection with GA
    checkpoint = load_checkpoint()
    if checkpoint and 'best_features' in checkpoint:
        selected_features = checkpoint['best_features']
    else:
        selected_features = genetic_algorithm_feature_selection(X_train, y_train, n_features=X_train.shape[1], n_selected_features=1000)

    X_train_selected = X_train[:, selected_features]
    X_test_selected = X_test[:, selected_features]
    print(f"Training with {len(selected_features)} selected features")

    # Save selected features
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/selected_features.pkl", 'wb') as f:
        pickle.dump(selected_features, f)

    # Define parameter grid for kNN
    param_grid = {
        'n_neighbors': [3, 5],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    }

    # Initialize kNN and GridSearchCV
    knn = KNeighborsClassifier()
    grid_search = CustomGridSearchCV(
        estimator=knn,
        param_grid=param_grid,
        scoring='accuracy',
        n_jobs=-1,
        cv=5,
        verbose=3,
        return_train_score=True
    )

    # Fit with checkpointing
    print("Starting GridSearchCV...")
    grid_search.fit(X_train_selected, y_train)

    # Print best parameters and score
    print(f"\nBest Parameters: {grid_search.best_params_}")
    print(f"Best Cross-Validation Accuracy: {grid_search.best_score_:.4f}")

    # Train final model with best parameters
    best_knn = KNeighborsClassifier(**grid_search.best_params_, n_jobs=1)  # Avoid parallelization in final model
    best_knn.fit(X_train_selected, y_train)

    # Evaluate on train and test sets
    y_train_pred = best_knn.predict(X_train_selected)
    y_test_pred = best_knn.predict(X_test_selected)

    train_metrics = compute_metrics(y_train, y_train_pred, "Training")
    test_metrics = compute_metrics(y_test, y_test_pred, "Testing")

    results = {
        'best_params': grid_search.best_params_,
        'best_cv_score': grid_search.best_score_,
        'selected_features': selected_features,
        'train_metrics': train_metrics,
        'test_metrics': test_metrics
    }
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/knn_ga_results.pkl", 'wb') as f:
        pickle.dump(results, f)

    return results

# Main execution
if __name__ == "__main__":
    global current_state
    current_state = {
        'iteration': 0,
        'ga_completed_generations': 0,
        'ga_population': [],
        'best_features': [],
        'tuning_state': {'iteration': 0, 'completed_params': [],
                         'cv_results': {'params': [], 'mean_test_score': [], 'std_test_score': [],
                                        'mean_precision': [], 'mean_recall': [], 'mean_f1': []}}
    }
    X_train, y_train, X_test, y_test = load_and_concatenate_features()
    results = knn_ga_classifier(X_train, y_train, X_test, y_test)
```

```python
import numpy as np
import pandas as pd
import pickle
import os
import hashlib
import tempfile
import shutil
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from deap import base, creator, tools, algorithms
import optuna
from tqdm import tqdm
import signal
import sys

# Signal handler for graceful interruption
def signal_handler(sig, frame):
    print("\nInterruption detected. Saving current state...")
    save_checkpoint(current_state)
    sys.exit(0)

signal.signal(signal.SIGINT, signal_handler)

# Compute MD5 hash for file integrity
def compute_md5(file_path):
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()

# Save checkpoint with atomic write
def save_checkpoint(state):
    checkpoint_dir = "/content/drive/MyDrive/EEG_Project/Classification_Results/Checkpoints/"
    os.makedirs(checkpoint_dir, exist_ok=True)
    temp_fd, temp_path = tempfile.mkstemp(suffix='.pkl', dir=checkpoint_dir)

    try:
        with os.fdopen(temp_fd, 'wb') as f:
            pickle.dump(state, f)
        checkpoint_path = os.path.join(checkpoint_dir, f"rf_ga_optuna_checkpoint_iter_{state.get('iteration', 0)}.pkl")
        shutil.move(temp_path, checkpoint_path)
        print(f"Checkpoint saved at iteration {state.get('iteration', 0)}: {checkpoint_path}")
    except Exception as e:
        print(f"Error saving checkpoint: {e}")
        if os.path.exists(temp_path):
            os.remove(temp_path)
        raise

# Load checkpoint with integrity check
def load_checkpoint():
    checkpoint_dir = "/content/drive/MyDrive/EEG_Project/Classification_Results/Checkpoints/"
    if not os.path.exists(checkpoint_dir):
        return None
    checkpoint_files = [f for f in os.listdir(checkpoint_dir) if f.startswith("rf_ga_optuna_checkpoint_iter_")]
    if not checkpoint_files:
        return None

    latest_checkpoint = max(checkpoint_files, key=lambda x: int(x.split('_iter_')[1].split('.pkl')[0]))
    checkpoint_path = os.path.join(checkpoint_dir, latest_checkpoint)

    try:
        md5_before = compute_md5(checkpoint_path)
        with open(checkpoint_path, 'rb') as f:
            state = pickle.load(f)
        md5_after = hashlib.md5(open(checkpoint_path, 'rb').read()).hexdigest()
        if md5_before != md5_after:
            print(f"Checkpoint corrupted: {checkpoint_path}")
            return None
        print(f"Loaded checkpoint: {checkpoint_path}")
        return state
    except Exception as e:
        print(f"Error loading checkpoint: {e}")
        return None

# Load and concatenate features without PCA
def load_and_concatenate_features():
    densenet_path = "/content/drive/MyDrive/EEG_Project/densenet169_features.pkl"
    efficientnet_path = "/content/drive/MyDrive/EEG_Project/efficientnetb3_features.pkl"

    with open(densenet_path, 'rb') as f:
        densenet_data = pickle.load(f)
    with open(efficientnet_path, 'rb') as f:
        efficientnet_data = pickle.load(f)

    if not isinstance(densenet_data, dict) or not isinstance(efficientnet_data, dict):
        raise ValueError(f"Expected dictionaries. Got DenseNet type: {type(densenet_data)}, EfficientNet type: {type(efficientnet_data)}")

    expected_samples = 12096
    densenet_keys = sorted(densenet_data.keys())
    efficientnet_keys = sorted(efficientnet_data.keys())

    if len(densenet_keys) != expected_samples or len(efficientnet_keys) != expected_samples:
        raise ValueError(f"Expected {expected_samples} samples, got {len(densenet_keys)} for DenseNet and {len(efficientnet_keys)} for EfficientNet")
    if densenet_keys != efficientnet_keys:
        raise ValueError("Image paths do not match")

    densenet_features = np.array([densenet_data[key] for key in densenet_keys])
    efficientnet_features = np.array([efficientnet_data[key] for key in efficientnet_keys])

    if densenet_features.shape != (12096, 1664):
        raise ValueError(f"Expected X_train shape (12096, 1664), got {densenet_features.shape}")
    if efficientnet_features.shape != (12096, 1536):
        raise ValueError(f"Expected X_test shape (12096, 1536), got {efficientnet_features.shape}")

    concatenated_features = np.concatenate((densenet_features, efficientnet_features), axis=1)
    print(f"Concatenated features shape: {concatenated_features.shape}")

    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/train_data.pkl", 'rb') as f:
        train_data = pickle.load(f)
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/test_data.pkl", 'rb') as f:
        test_data = pickle.load(f)

    y_train = np.array(train_data['y_train'])
    y_test = np.array(test_data['y_test'])

    if y_train.shape != (9648,) or y_test.shape != (2448,):
        raise ValueError(f"Expected y_train shape (9648,), y_test shape (2448,), got {y_train.shape}, {y_test.shape}")
    if not np.all(np.isin(y_train, [0, 1])) or not np.all(np.isin(y_test, [0, 1])):
        raise ValueError("Labels must be binary (0 or 1)")

    scaler = StandardScaler()
    X_train = scaler.fit_transform(concatenated_features[:9648])
    X_test = scaler.transform(concatenated_features[9648:12096])

    print(f"Training label distribution: {np.bincount(y_train)}")
    print(f"Testing label distribution: {np.bincount(y_test)}")

    return X_train, y_train, X_test, y_test

# Compute metrics and confusion matrix
def compute_metrics(y_true, y_pred, dataset_name, iteration=None, total_iterations=None, params=None):
    cm = confusion_matrix(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    if iteration is not None and total_iterations is not None:
        print(f"\nIteration {iteration}/{total_iterations} - Params: {params}, {dataset_name}:")
    else:
        print(f"\n{dataset_name} Metrics:")
    print(f"Confusion Matrix:\n{cm}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")

    return {
        'confusion_matrix': cm,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    }

# Genetic Algorithm for feature selection with detailed metrics
def evaluate_features(individual, X_train, y_train, cv, default_rf_params, gen, n_generations):
    selected_features = [i for i, bit in enumerate(individual) if bit]
    if not selected_features:
        return 0.0,

    X_train_selected = X_train[:, selected_features]
    rf = RandomForestClassifier(**default_rf_params)
    scores = []
    precisions = []
    recalls = []
    f1s = []

    for fold, (train_idx, val_idx) in enumerate(cv.split(X_train)):
        X_t, X_v = X_train_selected[train_idx], X_train_selected[val_idx]
        y_t, y_v = y_train[train_idx], y_train[val_idx]
        rf.fit(X_t, y_t)
        y_pred = rf.predict(X_v)
        metrics = compute_metrics(y_v, y_pred, f"Generation {gen+1} Fold {fold+1} Validation",
                                 iteration=f"{gen+1}.{fold+1}", total_iterations=f"{n_generations}.5",
                                 params=f"Features: {len(selected_features)}")
        scores.append(metrics['accuracy'])
        precisions.append(metrics['precision'])
        recalls.append(metrics['recall'])
        f1s.append(metrics['f1_score'])

    return np.mean(scores),

# Feature selection using Genetic Algorithm
def genetic_algorithm_feature_selection(X_train, y_train, n_features=3200, n_selected_features=1000, n_generations=10, population_size=20):
    global current_state
    checkpoint = load_checkpoint()
    if checkpoint and 'best_features' in checkpoint:
        print("Resuming with previously selected features...")
        return checkpoint['best_features']

    creator.create("FitnessMax", base.Fitness, weights=(1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMax)

    toolbox = base.Toolbox()
    toolbox.register("attr_bool", np.random.choice, [0, 1], p=[1 - n_selected_features/n_features, n_selected_features/n_features])
    toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, n=n_features)
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)
    toolbox.register("evaluate", evaluate_features, X_train=X_train, y_train=y_train,
                     cv=KFold(n_splits=5, shuffle=True, random_state=42),
                     default_rf_params={'n_estimators': 100, 'random_state': 42, 'n_jobs': -1},
                     n_generations=n_generations)
    toolbox.register("mate", tools.cxTwoPoint)
    toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
    toolbox.register("select", tools.selTournament, tournsize=3)

    population = toolbox.population(n=population_size)
    best_individual = None
    best_fitness = -float('inf')

    for gen in tqdm(range(n_generations), desc="GA Progress"):
        if checkpoint and 'ga_completed_generations' in checkpoint and gen < checkpoint['ga_completed_generations']:
            print(f"Skipping completed generation {gen+1}/{n_generations}")
            population = checkpoint['ga_population']
            continue

        offspring = algorithms.varAnd(population, toolbox, cxpb=0.5, mutpb=0.2)
        fits = [toolbox.evaluate(ind, gen=gen) for ind in offspring]

        for ind, fit in zip(offspring, fits):
            ind.fitness.values = fit

        population = toolbox.select(offspring, k=len(population))
        top_individual = tools.selBest(population, k=1)[0]
        top_fitness = top_individual.fitness.values[0]

        print(f"Generation {gen+1}/{n_generations} - Best Fitness (Accuracy): {top_fitness:.4f}, Features: {len([i for i, bit in enumerate(top_individual) if bit])}")

        if top_fitness > best_fitness:
            best_fitness = top_fitness
            best_individual = top_individual

        current_state['ga_completed_generations'] = gen + 1
        current_state['ga_population'] = population
        current_state['best_features'] = [i for i, bit in enumerate(best_individual) if bit]
        save_checkpoint(current_state)

    selected_features = [i for i, bit in enumerate(best_individual) if bit]
    print(f"Selected {len(selected_features)} features")
    return selected_features

# Optuna objective function for Random Forest hyperparameter tuning
def objective(trial, X_train, y_train, cv, class_weights):
    global current_state
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 200),
        'max_depth': trial.suggest_int('max_depth', 5, 50, log=True),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 4),
        'random_state': 42,
        'n_jobs': -1
    }

    trial_key = f"trial_{trial.number}"
    if trial_key in current_state['tuning_state']['completed_trials']:
        print(f"Skipping completed trial {trial.number}: {params}")
        return current_state['tuning_state']['cv_results'][trial_key]['mean_test_score']

    rf = RandomForestClassifier(**params, class_weight=class_weights)
    scores = []
    precisions = []
    recalls = []
    f1s = []

    for fold, (train_idx, val_idx) in enumerate(cv.split(X_train)):
        X_t, X_v = X_train[train_idx], X_train[val_idx]
        y_t, y_v = y_train[train_idx], y_train[val_idx]
        rf.fit(X_t, y_t)
        y_pred = rf.predict(X_v)
        metrics = compute_metrics(y_v, y_pred, f"Trial {trial.number+1} Fold {fold+1} Validation",
                                 iteration=f"{current_state['iteration']+1}/50", total_iterations="50",
                                 params=params)
        scores.append(metrics['accuracy'])
        precisions.append(metrics['precision'])
        recalls.append(metrics['recall'])
        f1s.append(metrics['f1_score'])

        trial.report(np.mean(scores), fold)
        if trial.should_prune():
            print(f"Pruning trial {trial.number} at fold {fold+1}")
            raise optuna.TrialPruned()

    current_state['iteration'] += 1
    current_state['tuning_state']['completed_trials'][trial_key] = {
        'params': params,
        'mean_test_score': np.mean(scores),
        'std_test_score': np.std(scores),
        'mean_precision': np.mean(precisions),
        'mean_recall': np.mean(recalls),
        'mean_f1': np.mean(f1s)
    }
    save_checkpoint(current_state)

    return np.mean(scores)

# Main classifier function with GA feature selection and Optuna tuning
def rf_ga_classifier(X_train, y_train, X_test, y_test):
    global current_state
    # Compute class weights to handle imbalance
    class_weights = {0: len(y_train) / (2 * np.bincount(y_train)[0]),
                     1: len(y_train) / (2 * np.bincount(y_train)[1])}

    # Feature selection with GA
    checkpoint = load_checkpoint()
    if checkpoint and 'best_features' in checkpoint:
        selected_features = checkpoint['best_features']
    else:
        selected_features = genetic_algorithm_feature_selection(X_train, y_train, n_features=X_train.shape[1], n_selected_features=1000)

    X_train_selected = X_train[:, selected_features]
    X_test_selected = X_test[:, selected_features]
    print(f"Training with {len(selected_features)} selected features")

    # Save selected features
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/selected_features_rf.pkl", 'wb') as f:
        pickle.dump(selected_features, f)

    # Optuna study for hyperparameter tuning
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    study = optuna.create_study(direction='maximize')
    n_trials = 50

    # Resume from checkpoint
    if checkpoint and 'tuning_state' in checkpoint and checkpoint['tuning_state']['completed_trials']:
        for trial_key, trial_data in checkpoint['tuning_state']['completed_trials'].items():
            trial = study.add_trial(
                params=trial_data['params'],
                distributions={
                    'n_estimators': optuna.distributions.IntDistribution(50, 200),
                    'max_depth': optuna.distributions.IntDistribution(5, 50, log=True),
                    'min_samples_split': optuna.distributions.IntDistribution(2, 10),
                    'min_samples_leaf': optuna.distributions.IntDistribution(1, 4)
                },
                value=trial_data['mean_test_score'],
                state=optuna.trial.TrialState.COMPLETE
            )

    # Optimize with progress bar
    for _ in tqdm(range(n_trials - len(study.trials)), desc="Optuna Trials"):
        study.optimize(lambda trial: objective(trial, X_train_selected, y_train, cv, class_weights), n_trials=1)

    # Print best parameters and score
    print(f"\nBest Parameters: {study.best_params}")
    print(f"Best Cross-Validation Accuracy: {study.best_value:.4f}")

    # Train final model with best parameters
    best_rf = RandomForestClassifier(**study.best_params, class_weight=class_weights, n_jobs=-1, random_state=42)
    best_rf.fit(X_train_selected, y_train)

    # Evaluate on train and test sets
    y_train_pred = best_rf.predict(X_train_selected)
    y_test_pred = best_rf.predict(X_test_selected)

    train_metrics = compute_metrics(y_train, y_train_pred, "Training")
    test_metrics = compute_metrics(y_test, y_test_pred, "Testing")

    results = {
        'best_params': study.best_params,
        'best_cv_score': study.best_value,
        'selected_features': selected_features,
        'train_metrics': train_metrics,
        'test_metrics': test_metrics
    }
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/rf_ga_optuna_results.pkl", 'wb') as f:
        pickle.dump(results, f)

    return results

# Main execution
if __name__ == "__main__":
    global current_state
    current_state = {
        'iteration': 0,
        'ga_completed_generations': 0,
        'ga_population': [],
        'best_features': [],
        'tuning_state': {'completed_trials': {}}
    }
    X_train, y_train, X_test, y_test = load_and_concatenate_features()
    results = rf_ga_classifier(X_train, y_train, X_test, y_test)
```

```python
import numpy as np
import pandas as pd
import pickle
import os
import hashlib
import tempfile
import shutil
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from sklearn.model_selection import KFold
from deap import base, creator, tools, algorithms
import optuna
from tqdm import tqdm
import signal
import sys

# Signal handler for graceful interruption
def signal_handler(sig, frame):
    print("\nInterruption detected. Saving current state...")
    save_checkpoint(current_state)
    sys.exit(0)

signal.signal(signal.SIGINT, signal_handler)

# Compute MD5 hash for file integrity
def compute_md5(file_path):
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()

# Save checkpoint with atomic write
def save_checkpoint(state):
    checkpoint_dir = "/content/drive/MyDrive/EEG_Project/Classification_Results/Checkpoints/"
    os.makedirs(checkpoint_dir, exist_ok=True)
    temp_fd, temp_path = tempfile.mkstemp(suffix='.pkl', dir=checkpoint_dir)

    try:
        with os.fdopen(temp_fd, 'wb') as f:
            pickle.dump(state, f)
        checkpoint_path = os.path.join(checkpoint_dir, f"xgb_ga_optuna_checkpoint_iter_{state.get('iteration', 0)}.pkl")
        shutil.move(temp_path, checkpoint_path)
        print(f"Checkpoint saved at iteration {state.get('iteration', 0)}: {checkpoint_path}")
    except Exception as e:
        print(f"Error saving checkpoint: {e}")
        if os.path.exists(temp_path):
            os.remove(temp_path)
        raise

# Load checkpoint with integrity check
def load_checkpoint():
    checkpoint_dir = "/content/drive/MyDrive/EEG_Project/Classification_Results/Checkpoints/"
    if not os.path.exists(checkpoint_dir):
        return None
    checkpoint_files = [f for f in os.listdir(checkpoint_dir) if f.startswith("xgb_ga_optuna_checkpoint_iter_")]
    if not checkpoint_files:
        return None

    latest_checkpoint = max(checkpoint_files, key=lambda x: int(x.split('_iter_')[1].split('.pkl')[0]))
    checkpoint_path = os.path.join(checkpoint_dir, latest_checkpoint)

    try:
        md5_before = compute_md5(checkpoint_path)
        with open(checkpoint_path, 'rb') as f:
            state = pickle.load(f)
        md5_after = hashlib.md5(open(checkpoint_path, 'rb').read()).hexdigest()
        if md5_before != md5_after:
            print(f"Checkpoint corrupted: {checkpoint_path}")
            return None
        print(f"Loaded checkpoint: {checkpoint_path}")
        return state
    except Exception as e:
        print(f"Error loading checkpoint: {e}")
        return None

# Load and concatenate features without PCA
def load_and_concatenate_features():
    densenet_path = "/content/drive/MyDrive/EEG_Project/densenet169_features.pkl"
    efficientnet_path = "/content/drive/MyDrive/EEG_Project/efficientnetb3_features.pkl"

    with open(densenet_path, 'rb') as f:
        densenet_data = pickle.load(f)
    with open(efficientnet_path, 'rb') as f:
        efficientnet_data = pickle.load(f)

    if not isinstance(densenet_data, dict) or not isinstance(efficientnet_data, dict):
        raise ValueError(f"Expected dictionaries. Got DenseNet type: {type(densenet_data)}, EfficientNet type: {type(efficientnet_data)}")

    expected_samples = 12096
    densenet_keys = sorted(densenet_data.keys())
    efficientnet_keys = sorted(efficientnet_data.keys())

    if len(densenet_keys) != expected_samples or len(efficientnet_keys) != expected_samples:
        raise ValueError(f"Expected {expected_samples} samples, got {len(densenet_keys)} for DenseNet and {len(efficientnet_keys)} for EfficientNet")
    if densenet_keys != efficientnet_keys:
        raise ValueError("Image paths do not match")

    densenet_features = np.array([densenet_data[key] for key in densenet_keys])
    efficientnet_features = np.array([efficientnet_data[key] for key in efficientnet_keys])

    if densenet_features.shape != (12096, 1664):
        raise ValueError(f"Expected X_train shape (12096, 1664), got {densenet_features.shape}")
    if efficientnet_features.shape != (12096, 1536):
        raise ValueError(f"Expected X_test shape (12096, 1536), got {efficientnet_features.shape}")

    concatenated_features = np.concatenate((densenet_features, efficientnet_features), axis=1)
    print(f"Concatenated features shape: {concatenated_features.shape}")

    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/train_data.pkl", 'rb') as f:
        train_data = pickle.load(f)
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/test_data.pkl", 'rb') as f:
        test_data = pickle.load(f)

    y_train = np.array(train_data['y_train'])
    y_test = np.array(test_data['y_test'])

    if y_train.shape != (9648,) or y_test.shape != (2448,):
        raise ValueError(f"Expected y_train shape (9648,), y_test shape (2448,), got {y_train.shape}, {y_test.shape}")
    if not np.all(np.isin(y_train, [0, 1])) or not np.all(np.isin(y_test, [0, 1])):
        raise ValueError("Labels must be binary (0 or 1)")

    scaler = StandardScaler()
    X_train = scaler.fit_transform(concatenated_features[:9648])
    X_test = scaler.transform(concatenated_features[9648:12096])

    print(f"Training label distribution: {np.bincount(y_train)}")
    print(f"Testing label distribution: {np.bincount(y_test)}")

    return X_train, y_train, X_test, y_test

# Compute metrics and confusion matrix
def compute_metrics(y_true, y_pred, dataset_name, iteration=None, total_iterations=None, params=None):
    cm = confusion_matrix(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    if iteration is not None and total_iterations is not None:
        print(f"\nIteration {iteration}/{total_iterations} - Params: {params}, {dataset_name}:")
    else:
        print(f"\n{dataset_name} Metrics:")
    print(f"Confusion Matrix:\n{cm}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")

    return {
        'confusion_matrix': cm,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    }

# Genetic Algorithm for feature selection with detailed metrics
def evaluate_features(individual, X_train, y_train, cv, model_params, gen, n_generations):
    selected_features = [i for i, bit in enumerate(individual) if bit]
    if not selected_features:
        return 0.0,

    X_train_selected = X_train[:, selected_features]
    scores = []
    precisions = []
    recalls = []
    f1s = []

    for fold, (train_idx, val_idx) in enumerate(cv.split(X_train)):
        X_t, X_v = X_train_selected[train_idx], X_train_selected[val_idx]
        y_t, y_v = y_train[train_idx], y_train[val_idx]
        dtrain = xgb.DMatrix(X_t, label=y_t)
        dval = xgb.DMatrix(X_v, label=y_v)
        evals = [(dval, 'validation')]
        model = xgb.train(
            params=model_params,
            dtrain=dtrain,
            num_boost_round=model_params.get('n_estimators', 100),
            evals=evals,
            early_stopping_rounds=10,
            verbose_eval=False
        )
        y_pred = (model.predict(dval) > 0.5).astype(int)
        metrics = compute_metrics(
            y_v,
            y_pred,
            f"Generation {gen+1} Fold {fold+1} Validation",
            iteration=f"{gen+1}.{fold+1}",
            total_iterations=f"{n_generations}.5",
            params=f"Features: {len(selected_features)}"
        )
        scores.append(metrics['accuracy'])
        precisions.append(metrics['precision'])
        recalls.append(metrics['recall'])
        f1s.append(metrics['f1_score'])

    return np.mean(scores),

# Feature selection using Genetic Algorithm
def genetic_algorithm_feature_selection(X_train, y_train, n_features=3200, n_selected_features=500, n_generations=5, population_size=20):
    global current_state
    checkpoint = load_checkpoint()
    if checkpoint and 'best_features' in checkpoint:
        print("Resuming with previously selected features...")
        return checkpoint['best_features']

    creator.create("FitnessMax", base.Fitness, weights=(1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMax)

    toolbox = base.Toolbox()
    toolbox.register("attr_bool", np.random.choice, [0, 1], p=[1 - n_selected_features/n_features, n_selected_features/n_features])
    toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, n=n_features)
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)
    toolbox.register("evaluate", evaluate_features, X_train=X_train, y_train=y_train,
                     cv=KFold(n_splits=5, shuffle=True, random_state=42),
                     model_params={'n_estimators': 100, 'max_depth': 6, 'learning_rate': 0.1, 'objective': 'binary:logistic', 'eval_metric': 'logloss', 'random_state': 42},
                     n_generations=n_generations)
    toolbox.register("mate", tools.cxTwoPoint)
    toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
    toolbox.register("select", tools.selTournament, tournsize=3)

    population = toolbox.population(n=population_size)
    best_individual = None
    best_fitness = -float('inf')

    for gen in tqdm(range(n_generations), desc="GA Progress"):
        if checkpoint and 'ga_completed_generations' in checkpoint and gen < checkpoint['ga_completed_generations']:
            print(f"Skipping completed generation {gen+1}/{n_generations}")
            population = checkpoint['ga_population']
            continue

        offspring = algorithms.varAnd(population, toolbox, cxpb=0.5, mutpb=0.2)
        fits = [toolbox.evaluate(ind, gen=gen) for ind in offspring]

        for ind, fit in zip(offspring, fits):
            ind.fitness.values = fit

        population = toolbox.select(offspring, k=len(population))
        top_individual = tools.selBest(population, k=1)[0]
        top_fitness = top_individual.fitness.values[0]

        print(f"Generation {gen+1}/{n_generations} - Best Fitness (Accuracy): {top_fitness:.4f}, Features: {len([i for i, bit in enumerate(top_individual) if bit])}")

        if top_fitness > best_fitness:
            best_fitness = top_fitness
            best_individual = top_individual

        current_state['ga_completed_generations'] = gen + 1
        current_state['ga_population'] = population
        current_state['best_features'] = [i for i, bit in enumerate(best_individual) if bit]
        save_checkpoint(current_state)

    selected_features = [i for i, bit in enumerate(best_individual) if bit]
    print(f"Selected {len(selected_features)} features")
    return selected_features

# Optuna objective function for XGBoost hyperparameter tuning
def objective(trial, X_train, y_train, cv, class_weights):
    global current_state
    trial_key = f"trial_xgb_{trial.number}"
    if trial_key in current_state['tuning_state']['completed_trials']:
        print(f"Skipping completed trial {trial.number}")
        return current_state['tuning_state']['completed_trials'][trial_key]['mean_test_score']

    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 150),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'random_state': 42,
        'scale_pos_weight': class_weights[1]
    }

    scores = []
    precisions = []
    recalls = []
    f1s = []

    for fold, (train_idx, val_idx) in enumerate(cv.split(X_train)):
        X_t, X_v = X_train[train_idx], X_train[val_idx]
        y_t, y_v = y_train[train_idx], y_train[val_idx]
        dtrain = xgb.DMatrix(X_t, label=y_t)
        dval = xgb.DMatrix(X_v, label=y_v)
        evals = [(dval, 'validation')]
        model = xgb.train(
            params=params,
            dtrain=dtrain,
            num_boost_round=params['n_estimators'],
            evals=evals,
            early_stopping_rounds=10,
            verbose_eval=False
        )
        y_pred = (model.predict(dval) > 0.5).astype(int)
        metrics = compute_metrics(
            y_v,
            y_pred,
            f"Trial {trial.number+1} Fold {fold+1} Validation",
            iteration=f"{current_state['iteration']+1}/30",
            total_iterations="30",
            params=params
        )
        scores.append(metrics['accuracy'])
        precisions.append(metrics['precision'])
        recalls.append(metrics['recall'])
        f1s.append(metrics['f1_score'])

        trial.report(np.mean(scores), fold)
        if trial.should_prune():
            print(f"Pruning trial {trial.number} at fold {fold+1}")
            raise optuna.TrialPruned()

    current_state['iteration'] += 1
    current_state['tuning_state']['completed_trials'][trial_key] = {
        'params': params,
        'mean_test_score': np.mean(scores),
        'std_test_score': np.std(scores),
        'mean_precision': np.mean(precisions),
        'mean_recall': np.mean(recalls),
        'mean_f1': np.mean(f1s)
    }
    save_checkpoint(current_state)

    return np.mean(scores)

# Main classifier function with GA feature selection and Optuna tuning
def xgb_ga_classifier(X_train, y_train, X_test, y_test):
    global current_state
    # Compute class weights to handle imbalance
    class_weights = {0: len(y_train) / (2 * np.bincount(y_train)[0]),
                     1: len(y_train) / (2 * np.bincount(y_train)[1])}

    # Feature selection with GA
    checkpoint = load_checkpoint()
    if checkpoint and 'best_features' in checkpoint:
        selected_features = checkpoint['best_features']
    else:
        selected_features = genetic_algorithm_feature_selection(
            X_train,
            y_train,
            n_features=X_train.shape[1],
            n_selected_features=500
        )

    X_train_selected = X_train[:, selected_features]
    X_test_selected = X_test[:, selected_features]
    print(f"Training with {len(selected_features)} selected features")

    # Save selected features
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/selected_features_xgb.pkl", 'wb') as f:
        pickle.dump(selected_features, f)

    # Optuna study for hyperparameter tuning
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    study = optuna.create_study(direction='maximize')
    n_trials = 30

    # Resume from checkpoint
    if checkpoint and 'tuning_state' in checkpoint and checkpoint['tuning_state']['completed_trials']:
        for trial_key, trial_data in checkpoint['tuning_state']['completed_trials'].items():
            trial = study.add_trial(
                params=trial_data['params'],
                distributions={
                    'n_estimators': optuna.distributions.IntDistribution(50, 150),
                    'max_depth': optuna.distributions.IntDistribution(3, 10),
                    'learning_rate': optuna.distributions.FloatDistribution(0.01, 0.2),
                    'subsample': optuna.distributions.FloatDistribution(0.5, 1.0),
                    'colsample_bytree': optuna.distributions.FloatDistribution(0.5, 1.0)
                },
                value=trial_data['mean_test_score'],
                state=optuna.trial.TrialState.COMPLETE
            )

    # Optimize with progress bar
    for _ in tqdm(range(n_trials - len(study.trials)), desc="Optuna Trials"):
        study.optimize(lambda trial: objective(trial, X_train_selected, y_train, cv, class_weights), n_trials=1)

    # Print best parameters and score
    print(f"\nBest Parameters: {study.best_params}")
    print(f"Best Cross-Validation Accuracy: {study.best_value:.4f}")

    # Train final model with best parameters
    best_params = study.best_params
    best_params.update({
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'random_state': 42,
        'scale_pos_weight': class_weights[1]
    })
    dtrain = xgb.DMatrix(X_train_selected, label=y_train)
    dtest = xgb.DMatrix(X_test_selected, label=y_test)
    evals = [(dtest, 'test')]
    best_model = xgb.train(
        params=best_params,
        dtrain=dtrain,
        num_boost_round=best_params['n_estimators'],
        evals=evals,
        early_stopping_rounds=10,
        verbose_eval=False
    )

    # Evaluate on train and test sets
    y_train_pred = (best_model.predict(dtrain) > 0.5).astype(int)
    y_test_pred = (best_model.predict(dtest) > 0.5).astype(int)

    train_metrics = compute_metrics(y_train, y_train_pred, "Training")
    test_metrics = compute_metrics(y_test, y_test_pred, "Testing")

    results = {
        'best_params': study.best_params,
        'best_cv_score': study.best_value,
        'selected_features': selected_features,
        'train_metrics': train_metrics,
        'test_metrics': test_metrics
    }
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/xgb_ga_optuna_results.pkl", 'wb') as f:
        pickle.dump(results, f)

    return results

# Main execution
if __name__ == "__main__":
    global current_state
    current_state = {
        'iteration': 0,
        'ga_completed_generations': 0,
        'ga_population': [],
        'best_features': [],
        'tuning_state': {'completed_trials': {}}
    }
    X_train, y_train, X_test, y_test = load_and_concatenate_features()
    results = xgb_ga_classifier(X_train, y_train, X_test, y_test)
```

```python
import numpy as np
import pandas as pd
import pickle
import os
import hashlib
import tempfile
import shutil
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from sklearn.model_selection import KFold
from deap import base, creator, tools, algorithms
import optuna
from tqdm import tqdm
import signal
import sys

# Signal handler for graceful interruption
def signal_handler(sig, frame):
    print("\nInterruption detected. Saving current state...")
    save_checkpoint(current_state)
    sys.exit(0)

signal.signal(signal.SIGINT, signal_handler)

# Compute MD5 hash for file integrity
def compute_md5(file_path):
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()

# Save checkpoint with atomic write
def save_checkpoint(state):
    checkpoint_dir = "/content/drive/MyDrive/EEG_Project/Classification_Results/Checkpoints_xgb/"
    os.makedirs(checkpoint_dir, exist_ok=True)
    temp_fd, temp_path = tempfile.mkstemp(suffix='.pkl', dir=checkpoint_dir)

    try:
        with os.fdopen(temp_fd, 'wb') as f:
            pickle.dump(state, f)
        checkpoint_path = os.path.join(checkpoint_dir, f"xgb_ga_optuna_checkpoint_iter_{state.get('iteration', 0)}.pkl")
        shutil.move(temp_path, checkpoint_path)
        print(f"Checkpoint saved at iteration {state.get('iteration', 0)}: {checkpoint_path}")
    except Exception as e:
        print(f"Error saving checkpoint: {e}")
        if os.path.exists(temp_path):
            os.remove(temp_path)
        raise

# Load checkpoint with integrity check
def load_checkpoint():
    checkpoint_dir = "/content/drive/MyDrive/EEG_Project/Classification_Results/Checkpoints_xgb/"
    if not os.path.exists(checkpoint_dir):
        return None
    checkpoint_files = [f for f in os.listdir(checkpoint_dir) if f.startswith("xgb_ga_optuna_checkpoint_iter_")]
    if not checkpoint_files:
        return None

    latest_checkpoint = max(checkpoint_files, key=lambda x: int(x.split('_iter_')[1].split('.pkl')[0]))
    checkpoint_path = os.path.join(checkpoint_dir, latest_checkpoint)

    try:
        md5_before = compute_md5(checkpoint_path)
        with open(checkpoint_path, 'rb') as f:
            state = pickle.load(f)
        md5_after = hashlib.md5(open(checkpoint_path, 'rb').read()).hexdigest()
        if md5_before != md5_after:
            print(f"Checkpoint corrupted: {checkpoint_path}")
            return None
        print(f"Loaded checkpoint: {checkpoint_path}")
        return state
    except Exception as e:
        print(f"Error loading checkpoint: {e}")
        return None

# Load and concatenate features without PCA
def load_and_concatenate_features():
    densenet_path = "/content/drive/MyDrive/EEG_Project/densenet169_features.pkl"
    efficientnet_path = "/content/drive/MyDrive/EEG_Project/efficientnetb3_features.pkl"

    with open(densenet_path, 'rb') as f:
        densenet_data = pickle.load(f)
    with open(efficientnet_path, 'rb') as f:
        efficientnet_data = pickle.load(f)

    if not isinstance(densenet_data, dict) or not isinstance(efficientnet_data, dict):
        raise ValueError(f"Expected dictionaries. Got DenseNet type: {type(densenet_data)}, EfficientNet type: {type(efficientnet_data)}")

    expected_samples = 12096
    densenet_keys = sorted(densenet_data.keys())
    efficientnet_keys = sorted(efficientnet_data.keys())

    if len(densenet_keys) != expected_samples or len(efficientnet_keys) != expected_samples:
        raise ValueError(f"Expected {expected_samples} samples, got {len(densenet_keys)} for DenseNet and {len(efficientnet_keys)} for EfficientNet")
    if densenet_keys != efficientnet_keys:
        raise ValueError("Image paths do not match")

    densenet_features = np.array([densenet_data[key] for key in densenet_keys])
    efficientnet_features = np.array([efficientnet_data[key] for key in efficientnet_keys])

    if densenet_features.shape != (12096, 1664):
        raise ValueError(f"Expected X_train shape (12096, 1664), got {densenet_features.shape}")
    if efficientnet_features.shape != (12096, 1536):
        raise ValueError(f"Expected X_test shape (12096, 1536), got {efficientnet_features.shape}")

    concatenated_features = np.concatenate((densenet_features, efficientnet_features), axis=1)
    print(f"Concatenated features shape: {concatenated_features.shape}")

    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/train_equal.pkl", 'rb') as f:
        train_data = pickle.load(f)
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/test_equal.pkl", 'rb') as f:
        test_data = pickle.load(f)

    y_train = np.array(train_data['y_train'])
    y_test = np.array(test_data['y_test'])

    if y_train.shape != (9360,) or y_test.shape != (2736,):
        raise ValueError(f"Expected y_train shape (9360,), y_test shape (2736,), got {y_train.shape}, {y_test.shape}")
    if not np.all(np.isin(y_train, [0, 1])) or not np.all(np.isin(y_test, [0, 1])):
        raise ValueError("Labels must be binary (0 or 1)")

    scaler = StandardScaler()
    X_train = scaler.fit_transform(concatenated_features[:9360])
    X_test = scaler.transform(concatenated_features[9360:12096])

    print(f"Training label distribution: {np.bincount(y_train)}")
    print(f"Testing label distribution: {np.bincount(y_test)}")

    return X_train, y_train, X_test, y_test

# Compute metrics and confusion matrix
def compute_metrics(y_true, y_pred, dataset_name, iteration=None, total_iterations=None, params=None):
    cm = confusion_matrix(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    if iteration is not None and total_iterations is not None:
        print(f"\nIteration {iteration}/{total_iterations} - Params: {params}, {dataset_name}:")
    else:
        print(f"\n{dataset_name} Metrics:")
    print(f"Confusion Matrix:\n{cm}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")

    return {
        'confusion_matrix': cm,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1
    }

# Genetic Algorithm for feature selection with detailed metrics
def evaluate_features(individual, X_train, y_train, cv, model_params, gen, n_generations):
    selected_features = [i for i, bit in enumerate(individual) if bit]
    if not selected_features:
        return 0.0,

    X_train_selected = X_train[:, selected_features]
    scores = []
    precisions = []
    recalls = []
    f1s = []

    for fold, (train_idx, val_idx) in enumerate(cv.split(X_train)):
        X_t, X_v = X_train_selected[train_idx], X_train_selected[val_idx]
        y_t, y_v = y_train[train_idx], y_train[val_idx]
        dtrain = xgb.DMatrix(X_t, label=y_t)
        dval = xgb.DMatrix(X_v, label=y_v)
        evals = [(dval, 'validation')]
        model = xgb.train(
            params=model_params,
            dtrain=dtrain,
            num_boost_round=model_params.get('n_estimators', 100),
            evals=evals,
            early_stopping_rounds=10,
            verbose_eval=False
        )
        y_pred = (model.predict(dval) > 0.5).astype(int)
        metrics = compute_metrics(
            y_v,
            y_pred,
            f"Generation {gen+1} Fold {fold+1} Validation",
            iteration=f"{gen+1}.{fold+1}",
            total_iterations=f"{n_generations}.5",
            params=f"Features: {len(selected_features)}"
        )
        scores.append(metrics['accuracy'])
        precisions.append(metrics['precision'])
        recalls.append(metrics['recall'])
        f1s.append(metrics['f1_score'])

    return np.mean(scores),

# Feature selection using Genetic Algorithm
def genetic_algorithm_feature_selection(X_train, y_train, n_features=3200, n_selected_features=500, n_generations=5, population_size=20):
    global current_state
    checkpoint = load_checkpoint()
    if checkpoint and 'best_features' in checkpoint:
        print("Resuming with previously selected features...")
        return checkpoint['best_features']

    creator.create("FitnessMax", base.Fitness, weights=(1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMax)

    toolbox = base.Toolbox()
    toolbox.register("attr_bool", np.random.choice, [0, 1], p=[1 - n_selected_features/n_features, n_selected_features/n_features])
    toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, n=n_features)
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)
    toolbox.register("evaluate", evaluate_features, X_train=X_train, y_train=y_train,
                     cv=KFold(n_splits=5, shuffle=True, random_state=42),
                     model_params={'n_estimators': 100, 'max_depth': 6, 'learning_rate': 0.1, 'objective': 'binary:logistic', 'eval_metric': 'logloss', 'random_state': 42},
                     n_generations=n_generations)
    toolbox.register("mate", tools.cxTwoPoint)
    toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
    toolbox.register("select", tools.selTournament, tournsize=3)

    population = toolbox.population(n=population_size)
    best_individual = None
    best_fitness = -float('inf')

    for gen in tqdm(range(n_generations), desc="GA Progress"):
        if checkpoint and 'ga_completed_generations' in checkpoint and gen < checkpoint['ga_completed_generations']:
            print(f"Skipping completed generation {gen+1}/{n_generations}")
            population = checkpoint['ga_population']
            continue

        offspring = algorithms.varAnd(population, toolbox, cxpb=0.5, mutpb=0.2)
        fits = [toolbox.evaluate(ind, gen=gen) for ind in offspring]

        for ind, fit in zip(offspring, fits):
            ind.fitness.values = fit

        population = toolbox.select(offspring, k=len(population))
        top_individual = tools.selBest(population, k=1)[0]
        top_fitness = top_individual.fitness.values[0]

        print(f"Generation {gen+1}/{n_generations} - Best Fitness (Accuracy): {top_fitness:.4f}, Features: {len([i for i, bit in enumerate(top_individual) if bit])}")

        if top_fitness > best_fitness:
            best_fitness = top_fitness
            best_individual = top_individual

        current_state['ga_completed_generations'] = gen + 1
        current_state['ga_population'] = population
        current_state['best_features'] = [i for i, bit in enumerate(best_individual) if bit]
        save_checkpoint(current_state)

    selected_features = [i for i, bit in enumerate(best_individual) if bit]
    print(f"Selected {len(selected_features)} features")
    return selected_features

# Optuna objective function for XGBoost hyperparameter tuning
def objective(trial, X_train, y_train, cv, class_weights):
    global current_state
    trial_key = f"trial_xgb_{trial.number}"
    if trial_key in current_state['tuning_state']['completed_trials']:
        print(f"Skipping completed trial {trial.number}")
        return current_state['tuning_state']['completed_trials'][trial_key]['mean_test_score']

    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 150),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'random_state': 42,
        'scale_pos_weight': class_weights[1]
    }

    scores = []
    precisions = []
    recalls = []
    f1s = []

    for fold, (train_idx, val_idx) in enumerate(cv.split(X_train)):
        X_t, X_v = X_train[train_idx], X_train[val_idx]
        y_t, y_v = y_train[train_idx], y_train[val_idx]
        dtrain = xgb.DMatrix(X_t, label=y_t)
        dval = xgb.DMatrix(X_v, label=y_v)
        evals = [(dval, 'validation')]
        model = xgb.train(
            params=params,
            dtrain=dtrain,
            num_boost_round=params['n_estimators'],
            evals=evals,
            early_stopping_rounds=10,
            verbose_eval=False
        )
        y_pred = (model.predict(dval) > 0.5).astype(int)
        metrics = compute_metrics(
            y_v,
            y_pred,
            f"Trial {trial.number+1} Fold {fold+1} Validation",
            iteration=f"{current_state['iteration']+1}/30",
            total_iterations="30",
            params=params
        )
        scores.append(metrics['accuracy'])
        precisions.append(metrics['precision'])
        recalls.append(metrics['recall'])
        f1s.append(metrics['f1_score'])

        trial.report(np.mean(scores), fold)
        if trial.should_prune():
            print(f"Pruning trial {trial.number} at fold {fold+1}")
            raise optuna.TrialPruned()

    current_state['iteration'] += 1
    current_state['tuning_state']['completed_trials'][trial_key] = {
        'params': params,
        'mean_test_score': np.mean(scores),
        'std_test_score': np.std(scores),
        'mean_precision': np.mean(precisions),
        'mean_recall': np.mean(recalls),
        'mean_f1': np.mean(f1s)
    }
    save_checkpoint(current_state)

    return np.mean(scores)

# Main classifier function with GA feature selection and Optuna tuning
def xgb_ga_classifier(X_train, y_train, X_test, y_test):
    global current_state
    # Compute class weights to handle imbalance
    class_weights = {0: len(y_train) / (2 * np.bincount(y_train)[0]),
                     1: len(y_train) / (2 * np.bincount(y_train)[1])}

    # Feature selection with GA
    checkpoint = load_checkpoint()
    if checkpoint and 'best_features' in checkpoint:
        selected_features = checkpoint['best_features']
    else:
        selected_features = genetic_algorithm_feature_selection(
            X_train,
            y_train,
            n_features=X_train.shape[1],
            n_selected_features=500
        )

    X_train_selected = X_train[:, selected_features]
    X_test_selected = X_test[:, selected_features]
    print(f"Training with {len(selected_features)} selected features")

    # Save selected features
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/selected_features_xgb.pkl", 'wb') as f:
        pickle.dump(selected_features, f)

    # Optuna study for hyperparameter tuning
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    study = optuna.create_study(direction='maximize')
    n_trials = 30

    # Resume from checkpoint
    if checkpoint and 'tuning_state' in checkpoint and checkpoint['tuning_state']['completed_trials']:
        for trial_key, trial_data in checkpoint['tuning_state']['completed_trials'].items():
            trial = study.add_trial(
                params=trial_data['params'],
                distributions={
                    'n_estimators': optuna.distributions.IntDistribution(50, 150),
                    'max_depth': optuna.distributions.IntDistribution(3, 10),
                    'learning_rate': optuna.distributions.FloatDistribution(0.01, 0.2),
                    'subsample': optuna.distributions.FloatDistribution(0.5, 1.0),
                    'colsample_bytree': optuna.distributions.FloatDistribution(0.5, 1.0)
                },
                value=trial_data['mean_test_score'],
                state=optuna.trial.TrialState.COMPLETE
            )

    # Optimize with progress bar
    for _ in tqdm(range(n_trials - len(study.trials)), desc="Optuna Trials"):
        study.optimize(lambda trial: objective(trial, X_train_selected, y_train, cv, class_weights), n_trials=1)

    # Print best parameters and score
    print(f"\nBest Parameters: {study.best_params}")
    print(f"Best Cross-Validation Accuracy: {study.best_value:.4f}")

    # Train final model with best parameters
    best_params = study.best_params
    best_params.update({
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'random_state': 42,
        'scale_pos_weight': class_weights[1]
    })
    dtrain = xgb.DMatrix(X_train_selected, label=y_train)
    dtest = xgb.DMatrix(X_test_selected, label=y_test)
    evals = [(dtest, 'test')]
    best_model = xgb.train(
        params=best_params,
        dtrain=dtrain,
        num_boost_round=best_params['n_estimators'],
        evals=evals,
        early_stopping_rounds=10,
        verbose_eval=False
    )

    # Evaluate on train and test sets
    y_train_pred = (best_model.predict(dtrain) > 0.5).astype(int)
    y_test_pred = (best_model.predict(dtest) > 0.5).astype(int)

    train_metrics = compute_metrics(y_train, y_train_pred, "Training")
    test_metrics = compute_metrics(y_test, y_test_pred, "Testing")

    results = {
        'best_params': study.best_params,
        'best_cv_score': study.best_value,
        'selected_features': selected_features,
        'train_metrics': train_metrics,
        'test_metrics': test_metrics
    }
    with open("/content/drive/MyDrive/EEG_Project/Classification_Results/xgb_ga_optuna_results.pkl", 'wb') as f:
        pickle.dump(results, f)

    return results

# Main execution
if __name__ == "__main__":
    global current_state
    current_state = {
        'iteration': 0,
        'ga_completed_generations': 0,
        'ga_population': [],
        'best_features': [],
        'tuning_state': {'completed_trials': {}}
    }
    X_train, y_train, X_test, y_test = load_and_concatenate_features()
    results = xgb_ga_classifier(X_train, y_train, X_test, y_test)
```
